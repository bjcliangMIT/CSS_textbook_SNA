# 佛罗伦萨与美第奇家族的崛起
说起文艺复兴就不得不提到十五世纪的佛罗伦萨，说起十五世纪的佛罗伦萨又不得不提到美第奇家族（Medici）。该家族在统治佛罗伦萨期间，资助了达芬奇、米开朗琪罗、拉斐尔等我们耳熟能详的艺术巨匠。但是，美第奇家族是如何崛起的，又是在怎样的环境下崛起的呢？在15世纪，佛罗伦萨的新晋显贵中许多都是银行家，他们与当权家族联姻，既融入了最高的统治阶层，也逐渐瓦解了古老政治家族对于权力的垄断，让“新贵”们逐渐进入统治集团。美第奇家族就是这些新晋家族的典型代表。事实上，直到14世纪晚期，控制佛罗伦萨的都是阿比奇（Albizzi）家族。不过由于在阿比奇领导下的佛罗伦萨频繁战败，该家族不得不开始依赖银行借贷，而美第奇家族坐拥当时欧洲最大的银行——美第奇银行，便借机登上了佛罗伦萨的政治舞台，逐渐威胁到了老贵族的统治地位。1433年，佛罗伦萨政治斗争到达白热化阶段，阿比奇家族流放了当时的美第奇家族长科西莫。然而，科西莫作为一位极善于施恩和拉拢人脉的政治家，次年便成功回归并最终取代了阿比奇的统治。

帕杰特对佛罗伦萨的家族兴衰史，从社会网络分析的角度进行了非常经典的研究。他通过几十年的档案学研究，基于佛罗伦萨那个时候的信件和文献，构建起佛罗伦萨家族之间的社会网络。他提出，在佛罗伦萨有三个主要的精英网络：政治、亲缘、和经济网络，也就是说，每个行动者有三个维度的网络连接（如下图所示）。他的理论是：当网络中的一个行动者将他在某一个网络中的局部结构转移到另一个网络中，就产生了组织性创新（organizational innovation），当这种创新在新的网络中产生外溢效应，生成了新的产业和领域，就产生了组织性发明（organizational invention）。

这个理论以佛罗伦萨为其中一个例子。一开始，即13世纪晚期到14世纪早期，欧洲大陆上的国际贸易公司主要以上层贵族家族为单位。但是在1378年佛罗伦萨的工人起义之后，作为一种缓和阶级矛盾的方式，佛罗伦萨的政治精英们将国内的银行家们“提拔”进议会，让他们加入了本由上层贵族家族把持的国际贸易领域。这些国内的银行家们更习惯遵循的不是那些古老的家族传承制度，而是师傅带徒弟、基于合同的合作模式。当他们把这样的师徒+合同逻辑带进来与原有的家族制逻辑合并，就创造出一种新的家族间的合伙制度（组织性创新）。这种合伙制度指的是公司以家族为单位，而公司间则建立基于合同和复式计帐的伙伴关系。这一制度促进公司开始围绕几个主要的赞助者（比如美第奇家族）形成更集中地金融化管理。于是，公司间贷款迅速增加，而更重要的是，它为美第奇家族通过金融手段控制其他公司及其背后的家族创造了条件。而且，这种新的组织逻辑逐渐以一种不可逆的趋势，通过金融家们的家庭和政治网络，逐渐渗入婚姻和政治领域并制度化（组织性发明）。家族之间的联姻开始比家族内部的继承更重要，人们在女儿嫁妆上花的钱甚至比给儿子继承的钱多。而科西莫，一个喜欢隐没在幕布之后，通过代理人操控婚姻、政治和经济网络的，“斯芬克斯一样”狡猾的政治家，得以最终带领美第奇家族获得统治地位。

帕杰特所使用的数据极为复杂，我们很难复现整个研究。本案例的主要背景是佛罗伦萨在大约1430年前后的政治斗争，尤其围绕着美第奇家族和斯特罗齐（Strozzi）家族。后者在1434年被流放之前，一直是美第奇家族最大的对手之一，也是佛罗伦萨最富裕的家族。数据可以从公开网站 下载，我们调用的是“padgett_data.xml”数据子集。在该数据中，共有两个行动者x行动者的（无向）网络，一个是PADGM婚姻网络，一个是PADGB商业网络。其中，婚姻网络记录的是家族之间的联姻，商业网络记录的是家族之间的金融往来（比如借贷和合作）。

## 导入数据和构建网络

```python
import numpy as np # numpy包可以帮忙简化数据处理的过程
import matplotlib.pyplot as plt # 为了作图
from collections import Counter # Counter(列表)可以用来统计列表中元素的频次
import networkx as nx # Python用来进行网络分析的包
import xml.etree.ElementTree as ET # 读入xml格式的原始文件
```

```python
tree = ET.parse('padgett_data.xml')
root = tree.getroot() 
root[0][1][0].attrib, root[0][1][1].attrib
```

将两个网络提取出来分别保存成两个邻边列表(edgelist)，并用这两个邻边列表创建两个无向网络nx.Graph()。

```python
padgm_edgelist = []
for child in root[0][1][0]: # PADGM
    temp = list(child.attrib.values())
# 这个数据收录了所有”可能“出现的边：其中确实存在的边权重为1，其实不存在的边的权重为0.
    if int(float(temp[3])) == 1: # 因此我们只需要保存确实存在的边
        padgm_edgelist.append((temp[0], temp[1])) 
```

```python
padgb_edgelist = []
for child in root[0][1][1]: # PADGB
    temp = list(child.attrib.values())
    if int(float(temp[3])) == 1:
        padgb_edgelist.append((temp[0], temp[1]))
```

```python
G_marriage = nx.Graph() # 创建一个新的空的无向图
G_marriage.add_edges_from(padgm_edgelist) # 从邻边列表构建图
nx.info(G_marriage) # 可以用info函数了解图的基本信息
```

```python
G_business = nx.Graph() 
G_business.add_edges_from(padgb_edgelist)
nx.info(G_business)
```

```python
# 在我们的数据里，商业网络的点集是婚姻网络的点集的子集，下列四个家族只存在于婚姻网络中
G_marriage.nodes() - G_business.nodes(), G_business.nodes() - G_marriage.nodes()
```

## 绘出婚姻和商业网络
在介绍具体函数之前，让我们先把两张网络图画出来，对它们有个大概的印象。

```python
# 如上所示，既然婚姻网络的点集更大，那么在这里我们用婚姻网络来计算点的位置
pos = nx.spring_layout(G_marriage, seed = 60)
```

```python
plt.figure(figsize=(10, 6))

#画边，alpha控制透明度。很多时候为了美观，我们会提高边的透明度（降低alpha）来突出点的信息
nx.draw_networkx_edges(G_marriage, pos, edge_color = "red", 
                       width = 1, alpha = 0.5) 
#为了美观，就不把点画出来了，仅把点代表的名字（label）标注出来
nx.draw_networkx_labels(G_marriage, pos)

plt.title('Marriage Network')
plt.axis('off'); plt.show()
```

```python
plt.figure(figsize=(10, 6))

nx.draw_networkx_edges(G_business, pos, edge_color = "green", 
                       width = 1, alpha = 0.5)
nx.draw_networkx_labels(G_business, pos)

plt.title('Business Network')
plt.axis('off'); plt.show()
```

## Networkx函数应用
现在，让我们以婚姻网络为例，了解一下基本的networkx函数运用。
#### 1. 获取图中点和边的信息

```python
G = G_marriage
print(G.nodes()) # 输出点集
```

```python
print(list(G.edges())[:3]) # 输出邻接列表
```

```python
A = nx.adjacency_matrix(G) # 输出邻接矩阵
print(A.todense())
```

#### 2. 邻居和点度
下面我们来考察斯特罗齐家族在佛罗伦萨婚姻网络中与其他家族的关联。

```python
# 列出所有该点在G中的邻居，即所有与STROZZI家族联姻的家族
list(G.neighbors('STROZZI')) 
```

在我们的数据中，有4个家族与STROZZI联姻，而有6个家族与MEDICI联姻了。

```python
print(G.degree['STROZZI']) # 用degree函数获取点度
print(G.degree(['STROZZI', 'MEDICI'])) 
```

#### 3. 中心度
那么斯特罗齐和美第奇两个家族在这个婚姻网络里的势力到底孰强孰弱？让我们来比较不同的中心度。首先，在中介中心度计算中，第一步是获取最短路径，当然也可单独获取从某一个点到另一个点的最短路径。如下所示，我们通过函数获取两点间的最短路径。结果表明，两大家族在这段时间内并没有什么直接的联姻，但是血缘关系也并不是太疏远，RIDOLFI家族或许在一定程度上充当了两者之间的协调人。

```python
nx.shortest_path(G, source='STROZZI', target='MEDICI')
```

```python
bet_cent = nx.algorithms.centrality.betweenness_centrality(G) # 计算中介中心度
sorted(bet_cent.items(), key = lambda x: -x[1])[:5] # 按中心度倒数排序，显示中心度前五的家族
```

虽然两大家族在联姻的数量上（度数）相差不大，但美第奇家族的中介中心度远比斯特罗齐高。我们接着计算四种主要的中心度（中介中心度、点度中心度、接近中心度、特征向量中心度），并绘制相应的柱状图。

```python
import seaborn as sns # 用于绘图
import networkx.algorithms.centrality as nx_cent # 计算中心度
# 为避免代码冗余，我们先建立一个方便遍历的列表
cent_list = [('Betweenness', nx_cent.betweenness_centrality),# 中介中心度
             ('Degree',      nx_cent.degree_centrality),# 点度中心度
             ('Closeness',   nx_cent.closeness_centrality), # 接近中心度
             ('Eigenvector', nx_cent.eigenvector_centrality)] # 特征向量中心度
```

```python
# 以2x2的格式绘制四张图，figsize控制图的总大小
# sharex的意思是共享x轴，即只保留最下面两张图的x轴
fig, axs = plt.subplots(2, 2, figsize = (10,6), sharex=True)

# enumerate函数将列表组合为一个索引序列，即pair为数据，i为其索引（数据下标）
for i, pair in enumerate(cent_list):
    # 获取画图的位置：
    # 比如说，列表中的第3项应该画在axs[3//2][3%2]=axs[1][1]即左下的位置
    ax = axs[i//2][i%2] 
    # 比如计算centrality=nx_cent.betweenness_centrality(G)
    centrality = pair[1](G) 
    # 然后把获得的centrality（格式是字典）按照中心度数值逆序排列，保存为一个列表
    centrality = sorted(centrality.items(), key = lambda x: -x[1])
    # 用seaborn包进行绘图 （用palette来选择色卡）
    sns.barplot(y=[i[0] for i in centrality], x=[i[1] for i in centrality], 
                palette="Blues_r", ax = ax)
    ax.set_xlim([0, 0.6]) # 确保x轴的比例尺是一致的
    ax.set_title(pair[0]) # 根据上面创建的列表命名每张图
    # 在每张图中把两个家族在Y轴标成红色
    [i.set_color('red') for i in ax.yaxis.get_ticklabels() 
                         if i.get_text() in ['MEDICI', 'STROZZI']]

fig.suptitle('Centrality') # 设置大标题
fig.tight_layout() # 让四张图没有重复的部分，排列得更好看
plt.show()
```

我们不难发现，两大家族在联姻的数量（点度中心度）上相差无几，而且均选择了相对来说是相似核心位置的家族（特征向量中心度）。但是，两者在中介中心度和接近中心度上差异明显。结合网络图来看，美第奇更大程度上控制了一些边缘家族（比如Pazzi）与这个上流贵族圈子联系的渠道（中介中心度），而且处于图的中心位置，平均而言更容易联系到其他的家族（接近中心度）。

综上，尽管我们并没有全部数据，但是从数据子集中，我们可以对1430年前后美第奇和斯特罗齐家族之间的政治斗争做一番（不一定准确的）推测。可以说，斯特罗齐家族虽然仍属显贵，但该家族对于这个婚姻网络的掌控力度抑或动员能力，已远不如美第奇家族。

#### 4. 全局数据
此外，我们还可以对全局网络进行一些简单的分析。结果显示，商业网络的全局集聚系数和网络平均集聚系数都比婚姻网络要高，因为商业网络里有更多的闭合三元组。这或许也呼应了商业合作和婚姻的不同性质：合作和信任容易基于相互引荐而传递到第三方，但是婚姻的联系却取决于更复杂的情感与血缘考量，一般难以直接传递。

```python
print('婚姻网络')
print("密度：", nx.classes.function.density(G_marriage))
print("全局集聚系数/传递性系数：", nx.algorithms.cluster.transitivity(G_marriage))
print("网络平均集聚系数：", nx.average_clustering(G_marriage))
```

```python
print('商业网络')
print("密度：", nx.classes.function.density(G_business))
print("全局集聚系数/传递性系数：", nx.algorithms.cluster.transitivity(G_business))
print("网络平均集聚系数：", nx.average_clustering(G_business))
```

#### 5. 可视化算法的比较
在以上分析中，我们一直利用spring_layout (Fruchterman-Reingold force-directed algorithm)算法生成各节点在结果图中的位置。实际上，我们也可以尝试调用其他有意思的Layout。以下代码中，Circular，Shell和Spiral都强制性地将节点按照给定的图案排布，Planar和Spring希望消除或者尽量减少边之间的重叠而让图尽量美观，而Spectral则强调画出点的聚类。

```python
# 为避免代码冗余，我们再次先建立一个方便遍历的列表
draw_list = [  
    ('Circular', nx.circular_layout), # 把节点排成环形
    ('Shell',    nx.shell_layout), # 将节点排成同心圆形状
    ('Spiral',   nx.spiral_layout), # 把节点排成螺旋形状
    ('Planar',   nx.planar_layout), # 确保边没有任何交叉，即将图"拍"成平面（无法变成平面则报错）
    ('Spring',   nx.spring_layout), # 一遍遍调整点的位置，令边的长度基本等同并且减少边之间的重叠
    ('Spectral', nx.spectral_layout), # 根据拉普拉斯矩阵进行聚类，并基于聚类作图
            ]

fig, axs = plt.subplots(2, 3, figsize = (15,8), sharex=True)

for i, pair in enumerate(draw_list):
    ax = axs[i//3][i%3]
    nx.draw(G_marriage, pos = pair[1](G_marriage), 
            node_size = 50, node_color = 'black', ax=ax)
    ax.set_title(pair[0])

plt.show()
```
